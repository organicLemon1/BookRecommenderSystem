<font size = 4> **POPULARITY BASED RECOMMENDER SYSTEM** </font>

```python
ratings_with_name = ratings.merge(books, on = 'ISBN')
```

```python
ratings_with_name.head()
```

```python
ratings_with_name.shape
```

```python
num_ratings_df = ratings_with_name.groupby('Book-Title')['Book-Rating'].count().reset_index()
num_ratings_df.head()
```

```python
num_ratings_df.rename(columns = {'Book-Rating':'Num-Ratings'}, inplace = True)
num_ratings_df.head()
```

```python
avg_ratings_df = ratings_with_name.groupby('Book-Title')['Book-Rating'].mean().round(2).reset_index()
avg_ratings_df.rename(columns = {'Book-Rating':'Avg-Rating'}, inplace = True)
avg_ratings_df.head()
```

```python
print(num_ratings_df.shape)
print(avg_ratings_df.shape)
```

```python
popularity_df= num_ratings_df.merge(avg_ratings_df, on = 'Book-Title')
popularity_df.head()
```

```python
popularity_df = popularity_df[popularity_df['Num-Ratings'] >= 250].sort_values('Avg-Rating', ascending = False).head(50)
popularity_df
```

```python
popularity_df = popularity_df.merge(books, on = 'Book-Title').drop_duplicates('Book-Title')[['Book-Title', 'Book-Author', 'Image-URL-L', 'Num-Ratings', 'Avg-Rating']]
popularity_df
```

<font size = 4> **COLLABORATIVE FILTERING BASED RECOMMENDER SYSTEM** </font>

```python
ratings_with_name.groupby('User-ID')['Book-Rating'].count()
```

```python
x = ratings_with_name.groupby('User-ID')['Book-Rating'].count() > 200
x[x]
```

```python
padhe_likhe_users = x[x].index
filtered_ratings = ratings_with_name[ratings_with_name['User-ID'].isin(padhe_likhe_users)]
filtered_ratings
```

```python
y = filtered_ratings.groupby('Book-Title')['Book-Rating'].count() >= 50
famous_books = y[y].index
final_ratings = filtered_ratings[filtered_ratings['Book-Title'].isin(famous_books)]
final_ratings
```

```python
final_ratings.drop_duplicates().shape
```

```python
pivot_Table = final_ratings.pivot_table(index = 'Book-Title', columns = 'User-ID', values = 'Book-Rating')
pivot_Table
```

```python
pivot_Table.fillna(0, inplace = True)
pivot_Table.head()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity_score = cosine_similarity(pivot_Table)
```

```python
similarity_score.shape
```

```python
similarity_score
```

```python
similarity_score[1]
```

```python
np.where(pivot_Table.index == '1st to Die: A Novel')[0][0]
```

```python
sorted(list(enumerate(similarity_score[1])), key = lambda x:x[1], reverse = True)
```

```python
def recommend(book_name):
    index = np.where(pivot_Table.index == book_name)[0][0]
    similar_items = sorted(list(enumerate(similarity_score[index])), key = lambda x:x[1], reverse = True)[1:6]
    data = []
    for i in similar_items:
        item = []
        temp_df = books[books['Book-Title'] == pivot_Table.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Author'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Image-URL-L'].values))
        data.append(item)
    return data
```

```python
recommend('The Da Vinci Code')
```
